### 👨‍🏫 Operator precedence (basically BODMAS)

BODMAS is a rule used to determine the correct order of operations in mathematics. 

BODMAS stands for:

B – Brackets
O – Orders (powers, exponents, roots)
D – Division
M – Multiplication
A – Addition
S – Subtraction

## Order to follow
1. Solve anything inside Brackets first.
2. Calculate Orders (e.g., squares, cubes, square roots).
3. Perform Division and Multiplication from left to right (whichever comes first).
4. Perform Addition and Subtraction from left to right (whichever comes first).

Important: Division is not always before multiplication, and addition is not always before subtraction. They have the same precedence, so evaluate them from left to right.

```python
2 + 2 - 3 = 4 - 3 = 1 >>>>>> following BODMAS
2 + 2 - 3 = 2 - 1 = 1

2 + 2 * 3 = 2 + 6 = 8 

2 * (2 + 3) = 2 * 5= 10
```

```python
print(2 + 2 * 3) ## 8
```
